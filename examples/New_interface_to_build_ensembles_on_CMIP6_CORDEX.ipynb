```python
#from climaf.api import *
import itertools

req_dict = dict(project='CMIP6',
                variable=['tas','pr'],
                table='Amon',
                experiment='rcp85',
                period=['2050','2070'],
                model='*',
                version='latest',
                first_available=[dict(realization='*'),
                                 dict(grid=['gn','gr','gr1','gr2']),
                                ]
                )



try:
    # -- Simplest case: req_dict points to a single dataset
    req = ds(**req_dict).explore('resolve')
    return req
except:
    # -- Otherwise: we treat the cases for ensembles
    ens_dict = dict()
    #
    if 'first_available' in req_dict:
        first_available = req_dict['first_available']
        req_dict.pop('first_available')
        first_available_req_dict = dict()
        for elt in first_available:
            first_available_req_dict[elt.keys()[0]] = '*'
    #
    # -- Get all the combinations for the user-provided values
    multiple_values_keys = dict()
    for key in req_dict:
        if isinstance(req_dict[key], list) or req_dict[key]=='*':
            multiple_values_keys[key] = list(req_dict[key])
    #
    keys = multiple_values_keys.keys()
    values = (multiple_values_keys[key] for key in keys)
    combinations = [dict(zip(keys, combination)) for combination in itertools.product(*values)]
    #
    #
    for combination in combinations:
        print 'combination = ',combination
        #
        # -- Build member name
        member_name = combination[sorted(combination)[0]]
        for elt in sorted(combination):
            if combination[elt] not in member_name:
                member_name = member_name+'_'+combination[elt]
            else:
                member_name = combination[elt]
        #
        tmp_req_dict = req_dict.copy()
        # -- Add the combination of values
        tmp_req_dict.update(combination)
        # -- allow all the possible values for first_available
        tmp_req_dict.update(first_available_req_dict)
        # -- Do the request
        tmp_req = ds(**tmp_req_dict)
        # -- Get the results
        choices_tmp_req = tmp_req.explore('choices')
        #
        multiple_values_keys = dict()
        for elt in choices_tmp_req:
            # -- Si l'attribut a plusieurs valeurs et que cet attribut est dans le dictionnaire
            #    de requete passe par l'utilisateur, alors on le garde
            if isinstance(choices_tmp_req[elt], list) and elt in tmp_req_dict:
                multiple_values_keys[elt] = choices_tmp_req[elt]
        # -- Si on a eu un seul attribut avec plusieurs valeurs, on utilise explore('ensemble')
        #    pour creer l'ensemble pour cette combination
        if len(multiple_values_keys)==0:
            print 'FYI: the ensembles are created only on attributes that are set to lists or "*" '
            ens_dict[member_name] = tmp_req.explore('resolve')
        # -- If we have one ensemble for one key
        if len(multiple_values_keys)==1:
            tmp_ens = tmp_req.explore('ensemble')
            for mem_tmp_ens in tmp_ens:
                ens_dict[mem_tmp_ens+'_'+member_name]=tmp_ens[mem_tmp_ens].explore('resolve')
        # -- If we have multiple values for different keys
        if len(multiple_values_keys)>1:
            # -- This is where we should take care of the 'first_available'
            # Idee: si les attributs a valeurs multiples font partie des first_available,
            # on construit des combinaisons de valeurs a tester dans l'ordre
            multiple_values_first_available = dict()
            for dict_first_available in first_available:
                if '*' in dict_first_available.values()[0]:
                    # -- Recuperer toutes les valeurs de la requete
                    multiple_values_first_available[dict_first_available.keys()[0]] = choices_tmp_req[dict_first_available.keys()[0]]
                else:
                    multiple_values_first_available[dict_first_available.keys()[0]] = dict_first_available.values()[0]
            # -- Build the combinations:
            ###########
            # -- Then try the combinations until you find one;
            # -- If not one reach a unique result, raise a warning
            # Si ce n'est pas le cas, met un warning pour prevenir qu'en cas d'erreur, il faudra preciser
            # les valeurs de cet attribut
            # Et on verra au moment du explore('resolve') si on trouve une valeur unique ou pas
            # si on a pas une valeur unique, on renvoie un message
            tmp_keys = multiple_values_keys.keys()
            tmp_values = (multiple_values_keys[key] for key in tmp_keys)
            tmp_combinations = [dict(zip(tmp_keys, tmp_combination)) for tmp_combination in itertools.product(*tmp_values)]
            for tmp_combination in tmp_combinations:
                ttmp_req_dict = tmp_req_dict.copy()
                ttmp_req_dict.update(tmp_combination)
                # member_name exists here
                for elt in sorted(tmp_combination):
                    if tmp_combination[elt] not in member_name:
                        member_name = member_name+'_'+tmp_combination[elt]
                #
                try:
                    ens_dict[member_name] = ds(**ttmp_req_dict).explore('resolve')
                except:
                    if ds(**ttmp_req_dict).explore('choices'):
                        print 'this request could not reach a unique result because of multiple possible values'
    #
    # -- Return the ensemble
    return cens(ens_dict)
```
